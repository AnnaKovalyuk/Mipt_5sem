```python
import numpy as np

import plotly
import plotly.graph_objects as go
import plotly.express as px

from numpy.linalg import inv
```

Исходная матрица

```python
N=100
a=20
A = np.zeros((N,N))
for j in range(N):
  for k in range(N):
    if j==k: A[j,k]=a
    elif j+1==k or j+2==k or j+3==k or j+4==k or j-1==k or j-2==k or j-3==k or j-4==k: A[j,k]=1


F = np.arange(1, 101)[:,np.newaxis]
print(A)
```

Метод Гаусса

```python
n = 100
a = np.concatenate((A,F), axis=1)
x = np.zeros(n)
# Forward stepping
for i in range(n):        
    for j in range(i+1, n):
        ratio = a[j,i]/a[i,i]
        
        for k in range(n+1):
            a[j,k] = a[j,k] - ratio*a[i,k]

# Back stepping and solutions
x[n-1] = a[n-1,n]/a[n-1,n-1]

for i in range(n-2,-1,-1):
    x[i] = a[i,n]
    
    for j in range(i+1,n):
        x[i] = x[i] - a[i,j]*x[j]
    
    x[i] = x[i]/a[i,i]

# Printing
print('\nРешение: ')
for i in range(n):
    print('X%d = %f' %(i,x[i]), end = '\t')
```

```python
F.T-np.matmul(A,x)
```

```python
def zeidel_iter(A, F, x):
    #A is a system's matrix (2D), and F is a right part vector (1D)
    n = len(A)
    x_new = np.copy(x)

    for i in range(n):
        s1 = sum(A[i][j] * x_new[j] for j in range(i))
        s2 = sum(A[i][j] * x[j] for j in range(i + 1, n))
        x_new[i] = (F[i] - s1 - s2) / A[i][i]

    return x_new
num =30
x = np.zeros(np.shape(A)[0])
for i in range(num):
    x = zeidel_iter(A, F, x)

print('\nSolution is: ')
for i in range(len(A)):
    print('X%d = %f' %(i,x[i]), end = '\t')
```

```python
#Невязки
F.T-np.matmul(A,x)
```

Собственные значения

```python
n = 100
a = np.copy(A)
# Written only for non-zero diag elements
def eigval_reverse(a):
    n = 100
    e = np.eye(n)
    for i in range(n):        
       for j in (np.append(np.arange(i), np.arange(i+1,n))):
            ratio = a[j,i]/a[i,i]
            for k in range(n):
                a[j,k] = a[j,k] - ratio*a[i,k]
                e[j,k] = e[j,k] - ratio*e[i,k]

    return [np.diag(a), (e.T/np.diag(a)).T]
```
