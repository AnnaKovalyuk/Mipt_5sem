```python
def F(x):
    return (x**2-3)/(x+2)

def FindMin(f, a, b, e=1e-10):
    d = (a - b)/1e10
    
    while ((b - a) > e):
        c = (a + b)/2
        
        if (f(c - d) > f(c + d)):
            b = c + d
        else:
            a = c - d
            
    return (a + b)/2
    
```

```python
print('Точка минимума функции: ', FindMin(F, -2, 0))
print('Значение функции: ', F(FindMin(F, -2, 0)))
```
